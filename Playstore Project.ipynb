# Data cleaning

```python
# List of characters to remove
chars_to_remove = ['+', ',', '$']
# List of column names to clean
cols_to_clean = ['Installs', 'Price']
```

```python
apps['Installs']
```

```python
#apps['Installs_Update'] = apps['Installs'].apply(lambda Installs: Installs.replace(',', ''))
#apps['Installs_Update']
```

```python
# Loop for each column in cols_to_clean
for col in cols_to_clean:
    # Loop for each char in chars_to_remove
    for char in chars_to_remove:
        # Replace the character with an empty string
        apps[col] = apps[col].str.replace(char, '')
        #apps[col] = apps[col].apply(lambda x: x.replace('+', ''))
        
# Print a summary of the apps dataframe
print (apps.info())
```

## Size and Price of an App

```python
#checking for irregularities
apps['Size'].unique()
```

```python
#I found  irregularities like 'Varies with device', M,K. We'd be replacing Varies with device with np.nan, then we'd strip and replace M,K with '000' and '' respectively
apps['Size']=apps.Size.replace('Varies with device', np.nan)
apps['Size']=apps.Size.str.replace('M', '000') #all size values becomes kilobyte type
apps['Size']=apps.Size.str.replace('k', '')
#apps['Size']=apps.Size.str.replace('1,000+', 1000)
apps['Size']=apps['Size'].astype(float)
apps['Size'].dtype
```

```python
# Select rows where both 'Rating' and 'Size' values are present 
apps_with_size_and_rating_present =apps[~apps[['Rating', 'Size']].isnull().any(axis=1)]
apps_with_size_and_rating_present
```

```python
# Subset for categories with at least 250 apps
large_categories = apps_with_size_and_rating_present.groupby(['Category']).filter(lambda x: len(x) >= 250)
large_categories
```

```python
# Plot size vs. rating
plt1 = sns.jointplot(x = large_categories['Size'], y = large_categories['Rating'])
```

```python
large_categories['Price'].unique()
```

```python
large_categories['Price']=large_categories['Price'].astype('float')
large_categories['Price'].dtype
```

```python
# Select apps whose 'Type' is 'Paid'. I would be using the new dataframe that we cleared off the null values from
paid_apps = large_categories[large_categories['Type'] == 'Paid']
paid_apps
```

```python
# Plot price vs. rating
plt2 = sns.jointplot(x = paid_apps['Price'], y = paid_apps['Rating'])
```

```python
#Inference; We find that the majority of top rated apps (rating over 4) range from 2 MB to 20 MB. We also find that 
#the vast majority of apps price themselves under $10.
```

## Relation between app category and app price

```python
# Select a few popular app categories
popular_app_cats = apps[apps.Category.isin(['GAME', 'FAMILY', 'PHOTOGRAPHY',
                                            'MEDICAL', 'TOOLS', 'FINANCE',
                                            'LIFESTYLE','BUSINESS'])]
popular_app_cats
```

```python
popular_app_cats['Price'].unique()
```

```python
popular_app_cats['Price']=popular_app_cats['Price'].astype('float')
popular_app_cats['Price'].dtype
```

```python
# Examine the price trend by plotting Price vs Category

fig, ax = plt.subplots()
fig.set_size_inches(15, 8)


ax = sns.stripplot(x = popular_app_cats['Price'], y = popular_app_cats['Category'], jitter=True, linewidth=1)
ax.set_title('App pricing trend across categories')
plt.show()
```

```python
# Apps whose Price is greater than 200
apps_above_200 = popular_app_cats[popular_app_cats['Price'] >200]
apps_above_200
```

```python
apps_above_200[['Category', 'App', 'Price']]
```

```python
'''Inference: Different categories demand different price ranges. Some apps that are simple and used daily, like the calculator 
app, should probably be kept free. However, it would make sense to charge for a highly-specialized medical app that 
diagnoses diabetic patients. Below, we see that Medical and Family apps are the most expensive. Some medical apps 
extend even up to $80! All game apps are reasonably priced below $20.'''
```

## Filter out "junk" apps

```python
# Select apps priced below $100
apps_under_100 = popular_app_cats[popular_app_cats['Price'] <100]
apps_under_100
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(15, 8)

# Examine price vs category with the authentic apps (apps_under_100)
ax = sns.stripplot(x =apps_under_100['Category'], y =apps_under_100['Price'], data =apps_under_100 , jitter = True, linewidth = 1)
ax.set_title('App pricing trend across categories after filtering for junk apps')
plt.show()
```
